```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('lena.png', 0)

#cv2.imshow('img', img)
#b, g, r = cv2.split(img)
hist = cv2.calcHist([img], [0], None, [256], [0, 256])
plt.plot(hist)

#for c in [b, g, r]:
#    plt.hist(c.ravel(), 256, [0, 256])

cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
messi = cv2.imread('messi.jpg')
messi = cv2.cvtColor(messi, cv2.COLOR_BGR2GRAY)
face = messi[5:35, 130:170]
h, w = face.shape

res = cv2.matchTemplate(messi, face, cv2.TM_CCOEFF_NORMED)
threshold = 0.9
loc = np.where(res >= threshold)

for pt in zip(*loc[::-1]):
    cv2.rectangle(messi, pt, (pt[0] + w, pt[1] + h), (0, 0, 0), 1)

plt.imshow(messi, cmap='gray')
```
