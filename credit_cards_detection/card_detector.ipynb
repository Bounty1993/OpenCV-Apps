```python
import imutils
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('card.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.xticks([])
plt.yticks([]);
```

```python
img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
_, thresh = cv2.threshold(img, 150, 200, cv2.THRESH_BINARY_INV)
```

```python
plt.imshow(thresh, cmap='gray_r')
```

```python
conts = cv2.findContours(thresh.copy(), cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
conts = imutils.grab_contours(conts)
```
