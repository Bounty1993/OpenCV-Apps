```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('card.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
plt.imshow(img)
plt.xticks([])
plt.yticks([]);
```

```python
img = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
_, thresh = cv2.threshold(img, 150, 200, cv2.THRESH_BINARY)
```

```python
plt.imshow(thresh)
```
